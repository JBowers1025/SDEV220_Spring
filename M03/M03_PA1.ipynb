##### 10.2 Define a generator function called get_odds() that returns the odd numbers from range(10). Use a for loop to find and print the third value returned.

```python
def get_odds():
    odd_nums = [num for num in range(10) if num % 2 == 1]
    return odd_nums

odds = get_odds()
print(odds)
```

```python

x = 0
for x in odds:
    x += 1
    if x == 2:
        print(odds[x])
```

##### 10.3 Define a decorator called test that prints start when a function is called, and end when it finishes.

```python
def test(func):
    def start_end(*args):
        print("Function Start")
        result = func(*args)
        print("Function End")
        return result
    return start_end

@test
def my_name(name):
    print(f"My name is {name}.")
    return f"Completed"

returning = my_name("Jenn")
print(f"Program: {returning}")
```
