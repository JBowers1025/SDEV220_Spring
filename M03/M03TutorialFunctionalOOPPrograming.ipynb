```python
class Solution:
    def sort012(self, arr):
        self.arr = arr
        n = len(arr)
        i = 0
        
        for i in range(n):
            max_n = i
            for num in range(i + 1, n):
                if arr[num] < arr[max_n]:
                    max_n = num
            arr[max_n], arr[i] = arr[i], arr[max_n]

        
Solution()
```

```python
class Solution:
    def binarysearch(self, arr, k):
        # Code Here
        self.arr = arr
        self.k = k
        
        low = 0
        high = len(arr) - 1
        
        while low <= high:
            mid = low + (high - low) // 2
            
            if arr[mid] == k:
                return mid
            elif arr[mid] < k:
                low = mid + 1
            else:
                high = mid - 1
                
        return -1

Solution()
```
