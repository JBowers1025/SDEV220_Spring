##### 8.10
###### Use a list comprehension to make a list called even of the even numbers in range(10).

```python
evenList = []
for number in range(10):
    if number % 2 != 1:
        evenList.append(number)
        
print(evenList)
```

```python
evenList = list(range(0,10,2))
print(evenList)
```

##### 9.6
###### Make a multilevel dictionary called life. Use these strings for the topmost keys: 'animals', 'plants', and 'other'. Make the 'animals' key refer to another dictionary with the keys 'cats', 'octopi', and 'emus'. Make the 'cats' key refer to a list of strings with the values 'Henri', 'Grumpy', and 'Lucy'. Make all the other keys refer to empty dictionaries.

```python
kitties = ['Henri', 'Grumpy', 'Lucy']
animals = {'cats' : kitties, 'octopi': {},'emus' : {}}
life = {'animals' : animals, 'plants': {}, 'other' : {}}
print(life)
```

##### 9.10
######  Use a dictionary comprehension to create the dictionary squares. Use range(10) to return the keys, and use the square of each key as its value.

```python
squares = {key: key**2 for key in range(10)}
print(squares)
```

##### 9.11
###### Use a set comprehension to create the set odd from the odd numbers in range(10).

```python
odd_set = {num for num in range(10) if num % 2 == 1}
print(odd_set)
```

##### 9.14
######  Use zip() to make a dictionary called movies that pairs these lists: titles = ['Creature of Habit', 'Crewel Fate', 'Sharks On a Plane'] and plots = ['A nun turns into a monster', 'A haunted yarn shop', 'Check your exits'].

```python
titles = ['Creature of Habit', 'Crewel Fate', 'Sharks On a Plane']
plots = ['A nun turns into a monster', 'A haunted yarn shop', 'Check your exits']
movies = dict(zip(titles, plots))
print(movies)
```
